```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
algo=['KNN','SVC','RF','XGB','DT','SGD']
```

```python
years=['2018','2019','2020','2021','2022']
```

```python
knn_d={}
for i in years:
    knn_d[i]=None
print(knn_d)
```

```python
knn_vals=[2240,1618,2526,2510,1563]
c=0
for i in knn_d:
    knn_d[i]=knn_vals[c]
    c+=1
knn_d
```

```python
svc_d={}
for i in years:
    svc_d[i]=None
svc_vals=[6489,6740,8320,7686,4447]
c=0
for i in svc_d:
    svc_d[i]=svc_vals[c]
    c+=1
svc_d
```

```python
rf_d={}
for i in years:
    rf_d[i]=None
rf_vals=[15999,11652,18263,24516,9589]
c=0
for i in rf_d:
    rf_d[i]=rf_vals[c]
    c+=1
rf_d
```

```python
sgd_d={}
for i in years:
    sgd_d[i]=None
sgd_vals=[3772,2795,3111,6169,4227]
c=0
for i in sgd_d:
    sgd_d[i]=sgd_vals[c]
    c+=1
sgd_d
```

```python
fig,ax=plt.subplots(figsize=(9,7))
plt.plot(list(rf_d.keys()),list(rf_d.values()),label='Random Forest',linestyle='dotted')
plt.plot(list(sgd_d.keys()),list(sgd_d.values()),label='Stochastic Gradient Descent',linestyle='--')
plt.plot(list(dt_d.keys()),list(dt_d.values()),label='Decision Trees',linestyle='-')
plt.plot(list(xg_d.keys()),list(xg_d.values()),label='XgBoost Classifier',linestyle='solid')
plt.plot(list(svc_d.keys()),list(svc_d.values()),label='Support Vector Classifier',linestyle='dashdot')
plt.plot(list(knn_d.keys()),list(knn_d.values()),label='K Nearest Neighbours',linestyle='dashed')
plt.legend()
plt.show()
fig.savefig("Algo_Comparision_Graph_.jpg")
plt.close(fig)
```
